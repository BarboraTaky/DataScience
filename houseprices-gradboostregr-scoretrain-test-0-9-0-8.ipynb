# EDA(Feature engineering)- Null and missing values in in TRAIN nad TEST DS

- 4 solutions, how to work with Null and Missing Values:
- 1.) null values removed- cars = cars.dropna()
- 2.) null values set to 0 - cars['horsepower'] = cars['horsepower'].fillna(0)
- 3.) null values set to mean- cars['horsepower'] = cars['horsepower'].fillna(cars['horsepower'].mean())
- 4.) null values set to median- cars['horsepower'] = cars['horsepower'].fillna(cars['horsepower'].median())

# Machine Learning- modelling

```python
#Libraries for machine learning- supervised learning- all posible algoritms
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

#Machine learning 
MLA = [
    #LinearRegression(),
    #DecisionTreeRegressor(),
    #RandomForestRegressor(),
    GradientBoostingRegressor(n_estimators= 85, max_depth = 5, learning_rate = 0.1), #the best results(metrics below) for GradientBoostingRegressor 
    #SVR(kernel = "rbf", C = 100, gamma = 0.1, epsilon = 1), #better results for SVR changing these paramters
]


from sklearn.model_selection import cross_val_score, StratifiedKFold #mělo by být co nejvíce vypovídající cross_val_score než další metriky
from sklearn.metrics import mean_squared_error

skf = StratifiedKFold(n_splits = 5)
```

## Split TRAIN data set- train, test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
data_train_no_object.describe()
```

## Check validation metrics

```python
for a in MLA:
  a.fit(X_train, y_train) #fitem hledám paramtry téta 0,1,..n té mojí rovnice, kdežto predictem dosazuji konrétní hodnoty x_test a vyjde mi y_predict
  scores = cross_val_score(a, X_train, y_train, cv = 5) #mělo by být co nejvíce vypovídající, Kfold má výhodu shuffle(promíchá data set)
  accu_train = a.score(X_train, y_train)
  accu_test = a.score(X_test, y_test)
  rmse = mean_squared_error(y_test, a.predict(X_test), squared= False)
  print(a.__class__.__name__)
  print("CV score: ", scores.mean()) #tahle metrika by měla být více vypovídající než u train_score, 
  print("Train score: ", round(accu_train, 2))
  print("Test score: ", round(accu_test, 2))
  print("RMSE: ", round(rmse, 2))
  print("---------------")
```
